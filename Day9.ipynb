```python
import pandas as pd
# Load with error handling
df = pd.read_csv("broken-csv.txt", on_bad_lines="skip")
```

```python
# Use signup_date as index
df.set_index('signup_date', inplace=True)

# Reset back to default
df.reset_index(inplace=True)
df
```

```python
# Extract domain from email
df['domain'] = df['email'].map(lambda x: x.split('@')[1])
df
```

```python
# Convert wide format to long format
melted = pd.melt(df, id_vars=['id','name'], value_vars=['age','domain'])
print(melted.head())
```

```python
# Pivot table of age by domain
pivot = df.pivot_table(values='age', index='domain', aggfunc='mean')
print(pivot)
```
