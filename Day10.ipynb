```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("broken-csv.txt", on_bad_lines="skip")
```

```python
df['email'] = df['email'].str.split(',').str[0]
df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')  # critical fix
df['age'] = df['age'].fillna(df['age'].mean())
df.drop_duplicates(inplace=True)
df['name'] = df['name'].str.replace('"','')

# Step 3: Create helper columns
df['domain'] = df['email'].str.split('@').str[1]
df['year'] = df['signup_date'].dt.year   # now wor
```

```python
grouped = df.groupby(['domain','year'])['age'].mean().reset_index()
print("Average age per domain/year:\n", grouped)
```

```python
# Pivot table of average age by domain and year
pivot = pd.pivot_table(df, values='age', index='domain', columns='year', aggfunc='mean')
print(pivot)
```

```python
# Count of signups per domain per year
print(pd.crosstab(df['domain'], df['year']))
```

```python
# Example: merge with another dataset (like purchases)
purchases = pd.DataFrame({
    'id':[1,2,3,4],
    'purchase_amount':[200,150,300,400]
})
merged = pd.merge(df, purchases, on='id', how='left')
print(merged.head())
```
