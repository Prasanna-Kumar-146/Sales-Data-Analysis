```python
import pandas as pd
# Load with error handling
df = pd.read_csv("broken-csv.txt", on_bad_lines="skip")
df
```

```python
import matplotlib.pyplot as plt
```

```python
df['age'].plot(kind='bar')
plt.title("Age of Employees")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```

```python
df['age'].plot(kind='hist', bins=5, rwidth=0.8)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
df.groupby('signup_date').size().plot(kind='line')
plt.title("Signups Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

```python
df['domain'] = df['email'].str.split('@').str[1]
df['domain'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Email Domain Distribution")
plt.show()
```
