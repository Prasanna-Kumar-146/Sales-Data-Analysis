```python
import pandas as pd
# Load with error handling
df = pd.read_csv("broken-csv.txt", on_bad_lines="skip")
```

```python
# Fill missing ages with average
df['age'] = df['age'].fillna(df['age'].mean())
df
```

```python
# Convert signup_date to datetime, replace errors with NaT
df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')  
df
```

```python
# Keep only the first email if multiple are present
df['email'] = df['email'].str.split(',').str[0]
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df['name'] = df['name'].str.replace('"','')
df
```

```python
df['domain'] = df['email'].str.split('@').str[1]
df
```

```python
df['year'] = df['signup_date'].dt.year
df
```
